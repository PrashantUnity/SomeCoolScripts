```csharp
// Install the ScottPlot NuGet package
#r "nuget:ScottPlot"

// Setup a custom formatter to display plots as images
using Microsoft.DotNet.Interactive.Formatting; 

using System;
using System.Drawing;

var random = new Random();
```

```csharp
Formatter.Register(typeof(ScottPlot.Plot), (p, w) => 
    w.Write(((ScottPlot.Plot)p).GetImageHTML()), HtmlFormatter.MimeType);
```

```csharp
// create sample data
double[] dataX = new double[] { 1, 2, 3, 4, 5 };
double[] dataY = new double[] { 1, 4, 9, 16, 25 };

// plot the data
var plt = new ScottPlot.Plot(400, 300);
plt.AddScatter(dataX, dataY);

// display the plot
plt
```

```csharp
// create sample data
double[] values = { 778, 43, 283, 76, 184 };
string[] labels = { "C#", "JAVA", "Python", "F#", "PHP" };

// plot the data
ScottPlot.Plot plt = new(600, 400);
var pie = plt.AddPie(values);
pie.SliceLabels = labels;
pie.ShowLabels = true;

// show the plot
plt
```

```csharp
var g  = new int[][]
{
    new int[]{0,10},
    new int[]{3,18},
    new int[]{5,5},
    new int[]{6,11},
    new int[]{11,14},
    new int[]{13,1},
    new int[]{15,1},
    new int[]{17,4}
}; 

var coord = new HashSet<(int,int)>();

while(coord.Count<=g.Length *2)
{
    coord.Add((random.Next(1,20),random.Next(1,20)));
}

int i=0; 

var plt = new ScottPlot.Plot(600, 400);
var visited = new Dictionary<int,(int,int)>();
var coordLs = coord.ToArray(); 
int j=0;
var bubbleSize = 3;
var myBubblePlot = plt.AddBubblePlot();
foreach(var i in g)
{
    if(!visited.ContainsKey(i[0]))
    {
        visited.Add(i[0],coordLs[j++]);  
    }   
    if(!visited.ContainsKey(i[1]))
    {
        visited.Add(i[1],coordLs[j++]);
    }
     Color bubbleColor = Color.Aqua;
   myBubblePlot.Add(
        x: visited[i[0]].Item1,
        y: visited[i[0]].Item1,
        radius: bubbleSize,
        fillColor: bubbleColor,
        edgeColor: Color.Transparent,
        edgeWidth: 1
    );
}

plt
```
